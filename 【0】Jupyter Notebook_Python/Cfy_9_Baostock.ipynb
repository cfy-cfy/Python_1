```python
# （25）获取指定日期全部股票的日K线数据
import baostock as bs
import pandas as pd
def download_data(date):
    bs.login()

    # 获取指定日期的指数、股票数据
    stock_rs = bs.query_all_stock(date)
    stock_df = stock_rs.get_data()     # 直接获取数据并装为df，不每行每行进行
    data_df = pd.DataFrame()
    for code in stock_df["code"]:
        print("Downloading :" + code)
        k_rs = bs.query_history_k_data_plus(code, "date,code,open,high,low,close,preclose,volume,amount,adjustflag,turn,tradestatus,pctChg,peTTM,pbMRQ,psTTM,pcfNcfTTM,isST", 
                                            date, date,frequency="d",adjustflag="2")
        data_df = data_df.append(k_rs.get_data())
    bs.logout()
    data_df.to_csv(r"C:\Users\hp\Desktop\demo_assignDayData.csv", encoding="gbk", index=False)
if __name__ == '__main__':
    # 获取指定日期全部股票的日K线数据
    download_data("2021-02-02")
```
