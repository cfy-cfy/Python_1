```python
# 【0】tkinter 基础：Lable + Button + Entry + title + geometry + Radiobutton + Combobox + Checkbutton + messagebox`
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

app=tk.Tk()
app.title('请问你是谁？')   #——————  窗口标题
app.geometry('300x400')     #——————  窗口尺寸

lb=tk.Label(app,text='Your Mr Right is coming',width=20,height=10,fg='green')            #—————— Label 对象
lb.pack(side=tk.LEFT,padx=5) # 这里用 side 指定位置,padx 来指定边距

text=tk.Entry(app,width=10)                                #———————文本框
text.pack(side=tk.LEFT,padx=5)

bt1=tk.Button(app,text='go away',width=10,bg='pink',command=change_lable)                #—————— 按钮对象
bt1.pack()
bt2=tk.Button(app,text='吃饭',width=10,bg='pink',command=lambda:change_lab(text.get()))  #——————  按钮对象
bt2.pack()

radio_bt1=tk.Radiobutton(app,text='pretty ? ',value=1)     #——————— 添加单选按钮
radio_bt1.pack()
radio_bt2=tk.Radiobutton(app,text='not pretty ?' ,value=2)
radio_bt2.pack()

cb=ttk.Combobox(app,width=20)                               #——————— 下拉选择框
cb['values']=('1','2','3','4','5')
cb.pack()

check_bt1=tk.Checkbutton(app,text='pretty ? ')              #——————— 多选框
check_bt1.pack()
check_bt2=tk.Checkbutton(app,text='not pretty ? ')
check_bt2.pack()

# messagebox.showinfo("Good Good Study "," Day Day Up !!!!!")     #——————— show 弹窗
# messagebox.show("Good Good Study "," Day Day Up !!!!!")         
# messagebox.showerror("Good Good Study "," Day Day Up !!!!!")   
# messagebox.showwarning("Good Good Study "," Day Day Up !!!!!")  

# messagebox.askokcancel("你被变绿了？？","你被绿了？？")         #——————— ask 弹窗
# messagebox.askquestion("你被变绿了？？","你被绿了？？") 
# messagebox.askretrycancel("你被变绿了？？","你被绿了？？") 
# messagebox.askyesno("你被变绿了？？","你被绿了？？") 
# messagebox.askyesnocancel("你被变绿了？？","你被绿了？？") 

# ————————————————————————————————————— 命令按钮事项
def change_lable():
    lb.config(text='I am so sorry !',fg='black')

def change_lab(name):
    lb.config(text=f'{name},你妈叫你回家吃饭!!!',fg='green')
# ————————————————————————————————————————
tk.mainloop()
```

```python
# 【0】tkinter 基础：Text 
import tkinter as tk
from tkinter import messagebox

app=tk.Tk()
app.title('请问你是谁？')   #——————  窗口标题
app.geometry('300x400')     #——————  窗口尺寸

txt=tk.Text()
txt.pack()

bt1=tk.Button(app,text='go away',width=10,bg='pink',command=showmsg)  #—————— 按钮对象
bt1.pack()

# 其中 get 可以传入两参数，一个是想要获取的行数，一个是想要获取该行中的位置，使用 tk.END 可以获取全部 
def showmsg():
    messagebox.showinfo("哇哈哈",txt.get('1.0',tk.END))       # —————— 获取 Text 的内容
#     messagebox.showinfo("哇哈哈","香飘飘")

tk.mainloop()
```

```python
# 【0】tkinter 基础：frame 布局
import tkinter as tk
from tkinter import messagebox

app=tk.Tk()
app.title('请问你是谁？')   #——————  窗口标题
app.geometry('300x500')     #——————  窗口尺寸

f1=tk.Frame(master=app,height=10)
lb=tk.Label(f1,text='请输入你的姓名：',fg='black')
lb.pack(side=tk.LEFT)
text=tk.Entry(f1)
text.pack(side=tk.LEFT)
f1.pack(pady=5)   #—————— 每个 Frame 设置的间隔为 5

f2=tk.Frame(master=app,height=10)
radio_bt1=tk.Radiobutton(f2,text='pretty ? ',value=1)     #——————— 添加单选按钮
radio_bt1.pack(side=tk.LEFT,padx=10)
radio_bt2=tk.Radiobutton(f2,text='not pretty ?' ,value=2)
radio_bt2.pack(side=tk.LEFT,padx=10)
f2.pack(pady=5)

f3=tk.Frame(master=app,height=10)
cb=ttk.Combobox(f3,width=20)                               #——————— 下拉选择框
cb['values']=('1','2','3','4','5')
cb.pack(side=tk.LEFT)
f3.pack(pady=5)

f4=tk.Frame(master=app,height=10)
check_bt1=tk.Checkbutton(f4,text='pretty ? ')              #——————— 多选框
check_bt1.pack(side=tk.LEFT)
check_bt2=tk.Checkbutton(f4,text='not pretty ? ')
check_bt2.pack(side=tk.LEFT)
f4.pack(pady=5)

f5=tk.Frame(master=app,height=10)
txt=tk.Text(f5)
txt.pack()
bt1=tk.Button(f5,text='go away',width=10,bg='pink',command=showmsg)  #—————— 按钮对象
bt1.pack()
def showmsg():
    messagebox.showinfo("哇哈哈",txt.get('1.0',tk.END))       # —————— 获取 Text 的内容
f5.pack(pady=5)

tk.mainloop()
```

```python
# 【0】tkinter 基础：grid 布局
import tkinter as tk
from tkinter import messagebox

app=tk.Tk()
app.title('请问你是谁？')   #——————  窗口标题
app.geometry('300x600')     #——————  窗口尺寸
app.rowconfigure([0,1,2,3],minsize=50)

f1=tk.Frame(master=app,height=10)
f1.grid(column=0,row=2,sticky='w',padx=5,pady=5)  # 2行 0 列 左对齐
lb=tk.Label(f1,text='请输入你的姓名：',fg='black')
lb.pack(side=tk.LEFT)
text=tk.Entry(f1)
text.pack(side=tk.LEFT)

f2=tk.Frame(master=app,height=10)
f2.grid(column=0,row=1,sticky='w',padx=5,pady=5)  # 1行 0 列
radio_bt1=tk.Radiobutton(f2,text='pretty ? ',value=1)     #——————— 添加单选按钮
radio_bt1.pack(side=tk.LEFT,padx=10)
radio_bt2=tk.Radiobutton(f2,text='not pretty ?' ,value=2)
radio_bt2.pack(side=tk.LEFT,padx=10)

f3=tk.Frame(master=app,height=10)
f3.grid(column=0,row=3,sticky='w')  # 3行 0 列 左对齐
cb=ttk.Combobox(f3,width=20)                               #——————— 下拉选择框
cb['values']=('1','2','3','4','5')
cb.pack(side=tk.LEFT)

f4=tk.Frame(master=app,height=10)
f4.grid(column=0,row=0,sticky='w')  # 0行 0 列 左对齐
check_bt1=tk.Checkbutton(f4,text='pretty ? ')              #——————— 多选框
check_bt1.pack(side=tk.LEFT)
check_bt2=tk.Checkbutton(f4,text='not pretty ? ')
check_bt2.pack(side=tk.LEFT)

f5=tk.Frame(master=app,height=10)
f5.grid(column=0,row=4)    # 4行 0 列 左对齐
txt=tk.Text(f5)
txt.pack()
bt1=tk.Button(f5,text='go away',width=10,bg='pink',command=showmsg)  #—————— 按钮对象
bt1.pack()
def showmsg():
    messagebox.showinfo("哇哈哈",txt.get('1.0',tk.END))       # —————— 获取 Text 的内容

tk.mainloop()
```
