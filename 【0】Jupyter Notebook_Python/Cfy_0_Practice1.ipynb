```python
# 【7】@符号用做函数的修饰符:
'''
可以在模块或者类的定义层内对函数进行修饰，出现在函数定义的前一行，不允许和函数定义在同一行。
一个修饰符就是一个函数，它将被修饰的函数作为参数，并返回修饰后的同名函数或其他可调用的东西
（如果返回不是一个可调用的对象那么会报错）。
'''
def addspam(func):
    def new(*args):
       print("spam,spam,spam")
       return func(*args)
    return new

@addspam
def useful(a,b):
    print(a**2+b**2)
    
useful(4,3)
addspam(useful(4,3))
```
