```python
# （二）ANSI 兼容终端里，可以用彩色显示文本而不仅仅是黑白
'''
格式: echo -e "\033[字背景颜色;字体颜色m字符串\033[0m" 
例如: echo -e "\033[41;36m something here \033[0m" 

其中41的位置代表底色, 36的位置是代表字的颜色 
那些ascii code 是对颜色调用的始末——  \033[ ; m …… \033[0m 
字色              背景              颜色
---------------------------------------
30                40              黑色
31                41              紅色
32                42              綠色
33                43              黃色
34                44              藍色
35                45              紫紅色
36                46              青藍色
37                47              白色
===============================================ANSI控制码的说明 
\33[0m 关闭所有属性 
\33[1m 设置高亮度 
\33[4m 下划线 
\33[5m 闪烁 
\33[7m 反显 
\33[8m 消隐 
\33[30m -- \33[37m 设置前景色 
\33[40m -- \33[47m 设置背景色 
\33[nA 光标上移n行 
\33[nB 光标下移n行 
\33[nC 光标右移n行 
\33[nD 光标左移n行 
\33[y;xH设置光标位置 
\33[2J 清屏 
\33[K 清除从光标到行尾的内容 
\33[s 保存光标位置 
\33[u 恢复光标位置 
\33[?25l 隐藏光标 
\33[?25h 显示光标
'''
print( '\033[47;32m 可用 \033[0m')
print("\033[0;30;41m color!!! \033[0m Hello \n")
# 其中41的位置代表字的背景色, 30的位置是代表字的颜色，0 是字的一些特殊属性，
# 0代表默认关闭，一些其他属性如闪烁、下划线等。ascii code 是对颜色进行调用的。
print("\033[1;40;32m color!!! \033[0m hello \n")
print("*************************背景相同，字体颜色相同，属性不同 \n")
print("\033[0;30;41m color!!! \033[0m Hello \n")
print("\033[1;30;41m color!!! \033[0m Hello \n")
print("\033[4;30;41m color!!! \033[0m Hello \n")
print("\033[5;30;41m color!!! \033[0m Hello \n")
print("\033[7;30;41m color!!! \033[0m Hello \n")
print("\033[8;30;41m color!!! \033[0m Hello \n")
print("*************************背景相同，字体颜色不同 \n")
print("\033[1;30;41m color!!! \033[0m Hello \n")
print("\033[1;31;41m color!!! \033[0m Hello \n")
print("\033[1;32;41m color!!! \033[0m Hello \n")
print("\033[1;33;41m color!!! \033[0m Hello \n")
print("\033[1;34;41m color!!! \033[0m Hello \n")
print("\033[1;35;41m color!!! \033[0m Hello \n")
print("\033[1;36;41m color!!! \033[0m Hello \n")
print("\033[1;37;41m color!!! \033[0m Hello \n")

print("*************************字体颜色相同，背景不同 \n")
print("\033[0;30;40m color!!! \033[0m Hello \n");
print("\033[0;30;41m color!!! \033[0m Hello \n");
print("\033[0;30;42m color!!! \033[0m Hello \n");
print("\033[0;30;43m color!!! \033[0m Hello \n");
print("\033[0;30;44m color!!! \033[0m Hello \n");
print("\033[0;30;45m color!!! \033[0m Hello \n");
print("\033[0;30;46m color!!! \033[0m Hello \n");
print("\033[0;30;47m color!!! \033[0m Hello \n");
```
